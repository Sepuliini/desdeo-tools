```python
%matplotlib notebook
from desdeo_tools.scalarization import StomASF, PointMethodASF
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.rand(10000,2)
ideal = np.asarray((0,0))
nadir = np.asarray((1,1))
reference_point = np.asarray((0.5,0.9))
```

```python
sca1 = StomASF(ideal=ideal)
sca2 = PointMethodASF(ideal=ideal, nadir=nadir)
sca = [sca1, sca2]
```

```python
sca1data = sca1(data,reference_point=reference_point)
sca2data = sca2(data,reference_point=reference_point)
```

```python
f, ax = plt.subplots()
points = ax.scatter(x=data[:,0], y=data[:,1], c=scadata)
ax.scatter(x=reference_point[0], y=reference_point[1], c='r')
f.colorbar(points)
```
